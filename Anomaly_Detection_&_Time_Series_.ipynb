# **Question 1**  
# **What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

# **Answer 1**

## **1. Introduction to Anomaly Detection**

**Anomaly Detection** is a technique used to identify unusual patterns, events, or observations in data that do not conform to expected behavior. These unusual cases are called **anomalies**, **outliers**, or **exceptions**.

Anomalies often indicate:
- Fraudulent behavior  
- System failures  
- Cyber-attacks  
- Medical abnormalities  
- Sensor malfunctions  
- Unexpected patterns  

### **Why Anomaly Detection is important?**
It helps organizations detect:
- **Fraud** in banking transactions  
- **Intrusions** in network security  
- **Faults** in industrial machines  
- **Health issues** using patient vital signs  
- **Abnormal patterns** in time-series or sensor data  

Anomaly detection is widely used in:  
Finance, Cybersecurity, Healthcare, IoT, Retail, Engineering, and many other fields.

---

## **2. Types of Anomalies**

Anomalies mainly fall into **three categories**:

1. **Point Anomalies**  
2. **Contextual Anomalies**  
3. **Collective Anomalies**

Each of these represents a different style of abnormal behavior in data.

---

## **2.1 Point Anomalies**

### **Definition:**  
A **point anomaly** occurs when a *single individual data point* is significantly different from the rest of the data.

### **Characteristics:**
- Easiest type to detect  
- Occurs mostly in simple numerical datasets  
- Does not depend on time, season, or context  

### **Examples:**  
1. **Credit Card Fraud Detection:**  
   If normal transactions are between ₹200–₹1500, but suddenly a transaction of ₹1,00,000 occurs, it is a point anomaly.

2. **Temperature Sensor:**  
   A sensor normally reads **22–28°C**, but suddenly shows **100°C**.  
   This single reading is an anomaly.

3. **Exam Scores:**  
   A class scores between **30–80**, but one student scores **0**.  
   That 0 is a point anomaly.

---

## **2.2 Contextual Anomalies**

### **Definition:**  
A **contextual anomaly** depends on the *context* (time, season, location, user behavior).  
A value that is normal in one context may be abnormal in another.

### **Characteristics:**
- Mostly found in **time-series data**  
- Context is important (day vs. night, summer vs. winter)  
- Requires understanding of the environment  

### **Examples:**  
1. **Temperature Example:**  
   - 30°C in **summer** → normal  
   - 30°C in **winter** → anomalous  
   Context = season.

2. **Human Activity:**  
   - Walking at **3 PM** → normal  
   - Walking at **3 AM** inside an office → suspicious  
   Context = time of day.

3. **Website Traffic:**  
   - 500 visitors at 2 PM → normal  
   - 500 visitors at 3 AM → unusual  
   Context = time.

---

## **2.3 Collective Anomalies**

### **Definition:**  
A **collective anomaly** occurs when a *group of data points together* represents abnormal behavior, even if individual points appear normal.

### **Characteristics:**
- Occurs in sequential or grouped data  
- Individual points may not look abnormal  
- Pattern or sequence is abnormal  

### **Examples:**  
1. **Cyberattack Detection:**  
   A sudden burst of network packets (DoS attack).  
   Each packet may look normal, but their **collective pattern** is unusual.

2. **ECG Signal (Heart Monitoring):**  
   A small segment of heartbeats shows unusual rhythm.  
   Individually, beats look fine, but the pattern is abnormal.

3. **Machine Sensor Data:**  
   Repeated small vibration spikes that collectively indicate a machine fault.

---

## **3. Comparison of Anomaly Types**

| Type of Anomaly | Meaning | When It Occurs | Example |
|-----------------|---------|----------------|---------|
| **Point Anomaly** | Single unusual data point | Static datasets | A ₹1,00,000 transaction |
| **Contextual Anomaly** | Normal in one context but abnormal in another | Time-series, seasonal data | 30°C in winter |
| **Collective Anomaly** | Group of points is abnormal | Sequential/structured data | Network attack pattern |

---

## **4. Conclusion**

Anomaly Detection is a crucial technique in modern data-driven systems.  
Understanding the **three types of anomalies**—**point**, **contextual**, and **collective**—is essential for detecting different types of unusual behaviors in real-world applications.  

Each type provides unique insights and is useful in different environments, such as finance, healthcare, cybersecurity, and industrial systems.



# **Question 2**  
# **Compare Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) in terms of their approach and suitable use cases**

# **Answer 2**

## **1. Introduction**

Anomaly detection algorithms differ in how they identify unusual points in data.  
Here we compare **Isolation Forest**, **DBSCAN**, and **Local Outlier Factor (LOF)**—three widely used unsupervised anomaly detection methods.

Each method uses a unique approach and works best in specific types of datasets (high-dimensional, density-based, or local neighborhood-based data).

---

## **2. Isolation Forest**

### **Approach**
- Based on **random decision trees**.
- Anomalies are easier to isolate → they require **fewer splits** in the trees.
- Normal points need more splits because they lie in dense regions.
- Works by computing **isolation scores**.

### **Strengths**
- Works very well on **high-dimensional data**.
- **Fast**, scalable for large datasets.
- Does not require distance or density calculations.
- Works even when data has complex feature relationships.

### **Weaknesses**
- Not very effective on **density-based anomalies**.
- Performance decreases with small datasets.

### **Suitable Use Cases**
- Fraud detection (credit card transactions)
- Network intrusion detection
- Large-scale industrial or IoT anomaly detection
- High-dimensional datasets (≥20 features)

---

## **3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

### **Approach**
- A **density-based clustering algorithm**.
- Points in high-density regions form clusters.
- Points in low-density regions are labeled as **noise/anomalies**.
- Uses two parameters:  
  - **eps** (neighborhood radius)  
  - **min_samples** (minimum points to form a dense cluster)

### **Strengths**
- Detects anomalies as **points not belonging to any cluster**.
- Can find arbitrarily shaped clusters.
- Works well for spatial or density-based data.

### **Weaknesses**
- Struggles with **varying densities**.
- Sensitive to parameter selection (eps, min_samples).
- Does not scale well for high-dimensional datasets.

### **Suitable Use Cases**
- Geographical/spatial data anomaly detection
- GPS tracking irregularities
- Detecting noise in sensor data
- Cluster-based outlier detection in 2D/3D datasets

---

## **4. Local Outlier Factor (LOF)**

### **Approach**
- Measures the **local density** of each data point.
- A point is an anomaly if its local density is much lower than that of its neighbors.
- LOF calculates the **Local Outlier Score (LOF score)**.

### **Strengths**
- Excellent for detecting **local anomalies**.
- Works well when data has clusters with different densities.
- More flexible than DBSCAN in density-varying datasets.

### **Weaknesses**
- Computationally expensive (especially in large datasets).
- Not ideal for high-dimensional data.
- Sensitive to the number of neighbors (k).

### **Suitable Use Cases**
- Detecting anomalies within local regions
- Fraud detection in small datasets
- Identifying abnormal behavior in clustered data
- Time-series neighborhood anomaly detection

---

## **5. Side-by-Side Comparison Table**

| Feature / Algorithm | **Isolation Forest** | **DBSCAN** | **Local Outlier Factor (LOF)** |
|---------------------|----------------------|------------|---------------------------------|
| **Approach** | Random tree-based isolation | Density-based clustering | Local density comparison |
| **Type of Anomalies** | Global anomalies | Low-density points (noise) | Local anomalies |
| **Best For** | High-dimensional data | Spatial & density-based data | Varying-density clusters |
| **Scalability** | Excellent (linear) | Moderate | Poor for large datasets |
| **Parameter Sensitivity** | Low | High (eps, min_samples) | Medium (neighbors) |
| **Cluster Detection** | No | Yes | No |
| **Output** | Isolation score | Noise/cluster labels | LOF score |
| **Computational Cost** | Low | Medium | High |

---

## **6. Summary**

- **Isolation Forest** → Best for **large, high-dimensional datasets** and **global anomalies**.  
- **DBSCAN** → Best for **spatial/density anomalies** and detecting **noise outside clusters**.  
- **LOF** → Best for identifying **local anomalies** in data with **varying densities**.

Together, these methods cover a wide range of anomaly detection scenarios, and choosing the right one depends on data shape, size, density, and computational needs.



# **Question 3**  
# **What are the key components of a Time Series? Explain each with one example.**

# **Answer 3**

## **1. Introduction**

A **time series** is a sequence of data points collected or recorded at regular time intervals  
(e.g., hourly temperature, daily stock prices, monthly sales).

Time series data contains several underlying patterns.  
Understanding these patterns helps in forecasting future values accurately.

The **four key components** of a time series are:

1. **Trend**
2. **Seasonality**
3. **Cyclic Component**
4. **Irregular/Random Component**

---

# **2. Components of Time Series**

---

## **2.1 Trend Component**

### **Definition**
The **trend** represents the long-term direction of data over time.  
It can be **upward**, **downward**, or **stable**.

### **Characteristics**
- Shows long-term movement  
- Does not depend on seasons  
- Usually smooth and gradual  

### **Example**
- A company's revenue increasing steadily from **₹50,000 to ₹1,50,000 over 5 years**  
  → indicates an **upward trend**.

---

## **2.2 Seasonal Component**

### **Definition**
**Seasonality** refers to patterns that repeat at **regular intervals** (daily, weekly, monthly, or yearly).

### **Characteristics**
- Short-term and repeating  
- Caused by weather, festivals, human habits  
- Predictable and consistent

### **Example**
- Ice cream sales increase every **summer** and decrease every **winter**.  
  This repeating pattern is **seasonality**.

---

## **2.3 Cyclic Component**

### **Definition**
A **cyclic component** refers to long-term fluctuations caused by **business cycles**, **economic conditions**, or **social trends**.

### **Characteristics**
- Not regular or fixed like seasonality  
- Cycles last **years instead of months**  
- Peaks and troughs occur due to external forces (economy, market conditions)

### **Example**
- Real estate prices follow **economic cycles**:  
  boom → stability → recession → recovery.  
  These cycles may last **5–10 years**.

---

## **2.4 Irregular (Random) Component**

### **Definition**
The **irregular or random component** represents unpredictable variations in the data caused by unexpected events.

### **Characteristics**
- Non-systematic  
- Non-repeating  
- Caused by accidents, natural disasters, sudden market shifts

### **Example**
- A sudden drop in airline ticket sales due to **a cyclone** or **pandemic outbreak**.  
  This drop cannot be predicted by trend or seasonality.

---

# **3. Summary Table**

| Component | Meaning | Pattern | Example |
|----------|---------|---------|---------|
| **Trend** | Long-term direction | Gradual | Sales increasing over years |
| **Seasonality** | Repeating short-term pattern | Regular | High electricity use in summer |
| **Cyclic** | Long-term economic cycles | Not fixed | Market boom and recession |
| **Irregular** | Unpredictable fluctuations | Random | Drop in sales due to storm |

---

# **4. Conclusion**

Time series analysis is based on understanding these four core components.  
By separating data into **trend**, **seasonal**, **cyclic**, and **irregular** patterns, analysts can build accurate forecasting models and understand the true behavior of the system over time.



# **Question 4**  
# **Define Stationarity in Time Series. How can you test and transform a non-stationary series into a stationary one?**

# **Answer 4**
## **1. What is Stationarity?**

A **stationary time series** is a series whose **statistical properties remain constant over time**.

A time series is considered *stationary* if:

1. **Mean** is constant  
2. **Variance** is constant  
3. **Autocovariance** depends only on the lag, not time  
4. No trend or seasonal component exists  

Stationary data is essential because **most forecasting models (ARIMA, SARIMA, etc.) work only on stationary data**.

---

## **2. Why Stationarity is Important?**

- Makes the data predictable  
- Ensures model stability and reliability  
- Avoids misleading correlations  
- Helps in easier parameter estimation for forecasting models

Example:  
Daily temperature rising during summer is **not stationary**,  
but daily temperature differences may be stationary.

---

# **3. How to Test Stationarity?**

There are three main methods:

---

## **3.1 Visual Inspection (Graph-Based)**

Plot the time series and check for:
- Trends  
- Changing variance  
- Seasonality  

If trends or periodic patterns exist → **non-stationary**.

### **Example**
A rising stock market price trend indicates non-stationarity.

---

## **3.2 Summary Statistics Check**

Divide data into two parts  
(early period vs. later period)  
and compare:

- Mean  
- Variance  
- Standard deviation  

If they differ significantly → **non-stationary**.

---

## **3.3 Statistical Tests**

### **A. Augmented Dickey-Fuller (ADF) Test**
- Null hypothesis (**H0**): Series is **non-stationary**  
- If **p-value < 0.05** → reject H0 → **series is stationary**

### **B. KPSS Test**
- Null hypothesis (**H0**): Series is **stationary**  
- If **p-value < 0.05** → series is **non-stationary**

### **Why use both?**
- ADF checks for unit root  
- KPSS confirms stationarity  
Using both gives accurate results.

---

# **4. How to Transform a Non-Stationary Series into a Stationary One?**

---

## **4.1 Differencing**

Subtract current value from previous value:

\[
Y_t' = Y_t - Y_{t-1}
\]

This removes trend and makes mean constant.

### **Example**
Stock prices are non-stationary.  
But daily returns = price(t) – price(t−1) is usually stationary.

#### Types:
- **First-order differencing**
- **Second-order differencing** (if needed)

---

## **4.2 Log Transformation**

Used when variance increases with time.  
Reduces the impact of large values.

### **Example**
Electricity consumption grows exponentially.  
Applying log reduces variance.

\[
Y_t' = \log(Y_t)
\]

---

## **4.3 Seasonal Differencing**

Useful when seasonality exists.

\[
Y_t' = Y_t - Y_{t-m}
\]

Where **m** = seasonal period (e.g., 12 for monthly data).

### **Example**
Sales increase every December.  
Seasonal differencing removes seasonal effects.

---

## **4.4 Detrending**

Remove the trend using:
- Regression model  
- Moving average  
- Polynomial trend removal  

### **Example**
Fit a linear trend and subtract it from the data.

---

## **4.5 Box–Cox Transformation**

A power transformation to stabilize variance.

Formula:

\[
Y_t' = \frac{Y_t^\lambda - 1}{\lambda}
\]

Used when traditional log transform is not sufficient.

---

# **5. Summary Table**

| Method | Purpose | Removes | Example |
|--------|---------|---------|---------|
| **Differencing** | Make mean constant | Trends | Stock price differencing |
| **Seasonal Differencing** | Remove periodicity | Seasonality | Monthly sales |
| **Log Transform** | Stabilize variance | Exponential growth | Electricity usage |
| **Detrending** | Remove long term movement | Linear/Non-linear trend | GDP trend removal |
| **Box–Cox** | Stabilize variance | Non-linear variance | Population growth |

---

# **6. Conclusion**

A stationary time series is crucial for reliable and accurate forecasting.  
Non-stationary data can be tested using **ADF**, **KPSS**, and visual methods.  
It can be transformed into a stationary series using:

- **Differencing**
- **Log/Power transformations**
- **Seasonal adjustments**
- **Detrending**

By applying these techniques, complex time series become suitable for modeling and forecasting.



# **Question 5**  
# **Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application**

# **Answer 5**

## **1. Introduction**

Time series forecasting uses statistical models to understand patterns such as trend, seasonality, and autocorrelation.  
The models **AR, MA, ARIMA, SARIMA, and SARIMAX** are widely used for forecasting future values based on past observations.

Each model differs in structure and the type of data it is suitable for.

---

# **2. Models and Their Structures**

---

## **2.1 AR (AutoRegressive) Model**

### **Structure**
- Uses **past values** of the series to predict future values.
- Represented as **AR(p)**, where *p* = number of past observations (lags).
  
$$
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \epsilon_t
$$

### **Application**
- Works best when **past values influence future values**.
- Suitable for **stationary time series** without seasonality.

### **Example**
Forecasting stock returns using past 5-day returns → AR(5).

---

## **2.2 MA (Moving Average) Model**

### **Structure**
- Uses **past forecast errors (residuals)**, not past values.
- Represented as **MA(q)**.

$$
Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q} + \epsilon_t
$$


### **Application**
- Useful when shocks or anomalies affect the series.
- Suitable for **stationary data**.

### **Example**
Electricity demand influenced by past random fluctuations.

---

## **2.3 ARIMA Model**

### **Structure**
- Combines AR + MA + Differencing
- Represented as **ARIMA(p, d, q)**

Where:  
- **p** = AR terms  
- **d** = differencing to remove trend  
- **q** = MA terms  

$$
\text{ARIMA} = \text{AR} + \text{I (Integrated)} + \text{MA}
$$

### **Application**
- Best for **non-stationary time series** with **trend** but **no seasonality**.
- Most widely used for economic and business forecasting.

### **Example**
Monthly sales data showing upward trend → apply ARIMA(1,1,1).

---

## **2.4 SARIMA (Seasonal ARIMA)**

### **Structure**
- Extension of ARIMA that handles **seasonality**.
- Notation: **SARIMA(p, d, q)(P, D, Q)m**

Where:  
- (p, d, q) → non-seasonal components  
- (P, D, Q) → seasonal components  
- **m** → seasonal period (e.g., m = 12 for monthly data)

### **Application**
- Best for data with **seasonal patterns** (monthly, quarterly, weekly).

### **Example**
Ice cream sales showing yearly seasonality (12-month cycle).  
Model: SARIMA(1,1,1)(1,1,1)12

---

## **2.5 SARIMAX (Seasonal ARIMA with Exogenous Variables)**

### **Structure**
- SARIMA + external (exogenous) variables.
- Notation: **SARIMAX(p, d, q)(P, D, Q)m + X**

Where **X** = external factors.

### **Application**
- Used when **outside variables** affect the time series.
- Ideal for forecasting influenced by:
  - temperature  
  - promotions  
  - holidays  
  - weather  
  - economic indicators  

### **Example**
Predicting sales using seasonality + holiday promotions + temperature.

---

# **3. Comparison Table**

| Model | Full Name | Structure | Handles Trend? | Handles Seasonality? | External Variables? | Suitable For |
|-------|-----------|-----------|-----------------|----------------------|----------------------|---------------|
| **AR(p)** | AutoRegressive | Past values | No | No | No | Simple stationary data |
| **MA(q)** | Moving Average | Past errors | No | No | No | Random shock-dependent data |
| **ARIMA(p,d,q)** | AutoRegressive Integrated Moving Average | AR + differencing + MA | Yes (d) | No | No | Non-seasonal trending data |
| **SARIMA(p,d,q)(P,D,Q)m** | Seasonal ARIMA | ARIMA + seasonal components | Yes | Yes | No | Seasonal time series |
| **SARIMAX** | SARIMA + exogenous | Seasonal ARIMA + X | Yes | Yes | Yes | Seasonal series with external factors |

---

# **4. Summary**

- **AR model** uses past values.  
- **MA model** uses past errors.  
- **ARIMA** handles trend but not seasonality.  
- **SARIMA** handles both trend and seasonality.  
- **SARIMAX** extends SARIMA by adding **external predictors**, making it the most flexible model.

Each model is suitable for different patterns, and choosing the right one improves forecasting accuracy.



# Question 6: Time Series Analysis
# AirPassengers & NYC Taxi Far

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

%matplotlib inline

# -------------------------------
# Part A: AirPassengers Dataset
# -------------------------------

# Load dataset
air = get_rdataset('AirPassengers', 'datasets').data

# Create datetime index
air['Month'] = pd.date_range(start='1949-01', periods=len(air), freq='M')
air.set_index('Month', inplace=True)
air.rename(columns={'value':'Passengers'}, inplace=True)

# Plot original series
plt.figure(figsize=(12,5))
plt.plot(air['Passengers'], marker='o', color='blue')
plt.title('AirPassengers Time Series')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()

# Decompose time series
decomp_air = seasonal_decompose(air['Passengers'], model='multiplicative', period=12)
trend_air = decomp_air.trend
seasonal_air = decomp_air.seasonal
residual_air = decomp_air.resid

# Plot decomposed components
plt.figure(figsize=(12,10))
plt.subplot(411)
plt.plot(air['Passengers'], label='Original', color='blue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend_air, label='Trend', color='orange')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal_air, label='Seasonality', color='green')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual_air, label='Residual', color='red')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# -------------------------------
# Part B: NYC Taxi Fare Dataset (Synthetic)
# -------------------------------

# Create synthetic daily NYC taxi fare data (1 year)
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
# Base fare amount with weekly seasonality
fare_base = 5000 + 200*np.sin(2*np.pi*dates.dayofweek/7)
# Add random noise
fare_amount = fare_base + np.random.normal(0, 300, len(dates))

nyc_daily = pd.DataFrame({'Fare': fare_amount}, index=dates)

# Plot original series
plt.figure(figsize=(12,5))
plt.plot(nyc_daily['Fare'], color='blue')
plt.title('NYC Taxi Daily Fare Time Series (Synthetic)')
plt.xlabel('Date')
plt.ylabel('Total Fare Amount')
plt.grid(True)
plt.show()

# Decompose time series (weekly seasonality)
decomp_nyc = seasonal_decompose(nyc_daily['Fare'], model='additive', period=7)
trend_nyc = decomp_nyc.trend
seasonal_nyc = decomp_nyc.seasonal
residual_nyc = decomp_nyc.resid

# Plot decomposed components
plt.figure(figsize=(12,10))
plt.subplot(411)
plt.plot(nyc_daily['Fare'], label='Original', color='blue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend_nyc, label='Trend', color='orange')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal_nyc, label='Seasonality', color='green')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual_nyc, label='Residual', color='red')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```


# Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare)
# to detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python

from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

# Use the synthetic NYC Taxi Fare data from previous question
# nyc_daily dataframe has columns: 'Fare', index: 'Date'

# Step 2: Prepare data for Isolation Forest
# For 2D visualization, let's create a simple feature set:
# Feature 1: Fare amount, Feature 2: Rolling mean (window=7)
nyc_daily['Fare_RollMean'] = nyc_daily['Fare'].rolling(window=7, min_periods=1).mean()
X = nyc_daily[['Fare', 'Fare_RollMean']].values

# Step 3: Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
nyc_daily['Anomaly'] = iso_forest.fit_predict(X)  # -1 for anomaly, 1 for normal

# Separate normal vs anomalies
normal = nyc_daily[nyc_daily['Anomaly'] == 1]
anomalies = nyc_daily[nyc_daily['Anomaly'] == -1]

# Step 4: Plot 2D scatter plot
plt.figure(figsize=(12,6))
plt.scatter(normal['Fare'], normal['Fare_RollMean'], c='blue', label='Normal', alpha=0.6)
plt.scatter(anomalies['Fare'], anomalies['Fare_RollMean'], c='red', label='Anomaly', marker='x', s=100)
plt.title('Isolation Forest Anomaly Detection on NYC Taxi Fare')
plt.xlabel('Fare')
plt.ylabel('7-day Rolling Mean of Fare')
plt.legend()
plt.grid(True)
plt.show()
```

# Question 8: Train a SARIMA model on the monthly airline passengers dataset.
# Forecast the next 12 months and visualize the results.

```python

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.datasets import get_rdataset

%matplotlib inline


air = get_rdataset('AirPassengers', 'datasets').data
air['Month'] = pd.date_range(start='1949-01', periods=len(air), freq='M')
air.set_index('Month', inplace=True)
air.rename(columns={'value':'Passengers'}, inplace=True)


# Step 2: Train SARIMA Model

# SARIMA(p,d,q)(P,D,Q,s) parameters:
# After visual inspection or trial, common starting point: (1,1,1)(1,1,1,12)
sarima_model = SARIMAX(air['Passengers'],
                       order=(1,1,1),
                       seasonal_order=(1,1,1,12),
                       enforce_stationarity=False,
                       enforce_invertibility=False)

sarima_result = sarima_model.fit(disp=False)


# Step 3: Forecast Next 12 Months

forecast = sarima_result.get_forecast(steps=12)
forecast_index = pd.date_range(start=air.index[-1] + pd.offsets.MonthBegin(),
                               periods=12, freq='M')
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)
forecast_ci = forecast.conf_int()


# Step 4: Plot Results

plt.figure(figsize=(12,6))
plt.plot(air['Passengers'], label='Historical', color='blue')
plt.plot(forecast_series, label='Forecast', color='red', marker='o')
plt.fill_between(forecast_index,
                 forecast_ci['lower Passengers'],
                 forecast_ci['upper Passengers'],
                 color='pink', alpha=0.3)
plt.title('SARIMA Forecast of AirPassengers for Next 12 Months')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

%matplotlib inline

# Use the synthetic NYC Taxi Fare data from previous question
# nyc_daily dataframe has columns: 'Fare', index: 'Date'


nyc_daily['Fare_RollMean'] = nyc_daily['Fare'].rolling(window=7, min_periods=1).mean()
X = nyc_daily[['Fare', 'Fare_RollMean']].values

# Step 3: Fit LOF model
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
nyc_daily['LOF_Anomaly'] = lof.fit_predict(X)  # -1 for anomaly, 1 for normal

# Separate normal vs anomalies
normal = nyc_daily[nyc_daily['LOF_Anomaly'] == 1]
anomalies = nyc_daily[nyc_daily['LOF_Anomaly'] == -1]

# Step 4: Plot anomalies on 2D scatter plot
plt.figure(figsize=(12,6))
plt.scatter(normal['Fare'], normal['Fare_RollMean'], c='blue', label='Normal', alpha=0.6)
plt.scatter(anomalies['Fare'], anomalies['Fare_RollMean'], c='red', label='Anomaly', marker='x', s=100)
plt.title('Local Outlier Factor (LOF) Anomaly Detection on NYC Taxi Fare')
plt.xlabel('Fare')
plt.ylabel('7-day Rolling Mean of Fare')
plt.legend()
plt.grid(True)
plt.show()
```

# Question 10: Real-Time Energy Demand Forecasting and Anomaly Detection Workflow

You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.

---
# **Answer 10**
## 1. Detecting Anomalies in Streaming Data

For real-time anomaly detection:

- **Techniques:**
  - **Isolation Forest:** Efficient for high-dimensional streaming data; isolates anomalies without assuming a distribution.
  - **Local Outlier Factor (LOF):** Detects local density deviations; useful for contextual anomalies.
  - **DBSCAN:** Detects clusters; points outside clusters are anomalies (less suitable for real-time streaming).

- **Workflow in Streaming Context:**
  1. Maintain a **sliding window** of recent data (e.g., last 24 hours).
  2. Apply Isolation Forest or LOF on features (`energy usage`, `region`, `temperature`, `humidity`).
  3. Flag deviations as anomalies for **alerts** or **investigation**.
  ## 2. Short-Term Forecasting Model

For short-term forecasting, SARIMAX is recommended:

- Captures **seasonality** (daily patterns) and **trends**.
- Can include **exogenous variables** (temperature, humidity, region).
- Steps:
  1. Preprocess data into 15-minute intervals.
  2. Fit SARIMAX using historical energy usage + weather features.
  3. Forecast next few hours and continuously update as new data arrives.
## 3. Validation and Monitoring

- **Validation:**
  - Rolling-window cross-validation on historical data.
  - Metrics: MAE, RMSE, MAPE for forecasting; Precision/Recall for anomalies.

- **Real-Time Monitoring:**
  - Compare forecasted vs actual energy usage.
  - Track drift in model performance.
  - Retrain model if accuracy drops.
## 4. Business Impact

- **Operational Efficiency:** Detect abnormal spikes/drops to prevent blackouts or transformer overloads.
- **Cost Optimization:** Reduce overproduction or underutilization by accurate forecasts.
- **Customer Service & Safety:** Proactively manage critical areas during peaks or extreme weather.

---





```python

import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

%matplotlib inline

# Simulate streaming energy data (15-min intervals for 1 day)
np.random.seed(42)
timestamps = pd.date_range(start='2023-01-01', periods=96, freq='15T')
energy_usage = 100 + 10*np.sin(np.linspace(0, 6*np.pi, 96)) + np.random.normal(0, 5, 96)
energy_data = pd.DataFrame({'timestamp': timestamps, 'energy_usage': energy_usage})

# Apply Isolation Forest on the sliding window (full data for demonstration)
iso = IsolationForest(contamination=0.05, random_state=42)
energy_data['anomaly'] = iso.fit_predict(energy_data[['energy_usage']])

# Plot anomalies
plt.figure(figsize=(12,5))
plt.plot(energy_data['timestamp'], energy_data['energy_usage'], label='Energy Usage', color='blue')
plt.scatter(energy_data.loc[energy_data['anomaly']==-1, 'timestamp'],
            energy_data.loc[energy_data['anomaly']==-1, 'energy_usage'],
            color='red', marker='x', s=100, label='Anomaly')
plt.title('Real-Time Anomaly Detection (Isolation Forest)')
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage')
plt.legend()
plt.grid(True)
plt.show()
```
